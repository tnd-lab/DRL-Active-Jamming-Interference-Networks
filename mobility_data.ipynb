```python
from mobility_model_lib import *
import numpy as np
import pandas as pd 



NUMBER_OF_POSITIONS = 100
MAX_X = 500
MAX_Y = 500

# For Random Walk
# ---------------

NUMBER_OF_NODES_RW = 1 # jammer

# rw = random_walk(NUMBER_OF_NODES_RW, dimensions=(MAX_X, MAX_Y))

rw = random_waypoint(NUMBER_OF_NODES_RW, dimensions=(MAX_X, MAX_Y), velocity=(0.1, 1), wt_max=.1)
# This will create a Random Waypoint instance with 200 nodes in a simulation area of 100x100 units,
# velocity chosen from a uniform distribution between 0.1 and 1.0 units/step and maximum waiting time of 1.0 steps.
# https://github.com/panisson/pymobility#references

pos_rw_all = np.zeros([NUMBER_OF_POSITIONS, 2])


for idx in np.arange(NUMBER_OF_POSITIONS):
    pos = next(rw)
    pos_rw_all[idx] = pos
    if idx % 1000 == 0:
        print('Pos no. %.0f' % idx)

pd.DataFrame(pos_rw_all).to_csv("data_pos_RWP.csv", header = None, index = None)

# ---------------------------------------------------------------------
# Reference Point Group model

NUMBER_OF_NODES_RPGM = 12 # legitimate nodes and CCCH

rpgm = reference_point_group(NUMBER_OF_NODES_RPGM, dimensions=(MAX_X, MAX_Y), velocity=(0.1, 1), aggregation=0)


positions_all = np.zeros([NUMBER_OF_POSITIONS, 2 * NUMBER_OF_NODES_RPGM])


for idx in np.arange(NUMBER_OF_POSITIONS):
    positions = next(rpgm)
    positions_all[idx] = positions.reshape((1, 2 * NUMBER_OF_NODES_RPGM))
#
pd.DataFrame(positions_all).to_csv("data_pos_RPGM.csv", header=None, index=None)


```
